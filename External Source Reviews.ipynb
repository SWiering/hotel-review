# All reviews in a collection, to get all data in a unified way

```python
all_reviews = [reviews_1, reviews_2, reviews_3]
```

```python
reviews_list = []

for _page in all_reviews:
    _page_content = BeautifulSoup(_page.content)
    for review in _page_content.find_all('q'):
        reviews_list.append(review.getText())
        
reviews_list
```

```python
def call_procedure(function_name, params):
    
    connection = sqla.create_engine('mysql://root:@localhost/hotel-reviews?charset=utf8').raw_connection()
    try:
        cursor = connection.cursor()
        cursor.callproc(function_name, params)
        results = list(cursor.fetchall())
        cursor.close()
        connection.commit()
        return results
    finally:
        connection.close()
```

```python
call_procedure('addReview', [reviews_list[14], -1, 1])
```
